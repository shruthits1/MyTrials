```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("my_excel.xlsx")
```

```python
new = df[~(df['role'].isna())]
new
```

```python
d1 = {}
```

```python
for index, row in df.iterrows():
  if pd.isna(row["role"]) == False:
    if d1.keys().__contains__(row["sd_id"]) == False:
      d1[row["sd_id"]]={}
    print("processing..", row["sd_name"])
    d1[row["sd_id"]][row["role"]] = row["objectID"]
d1
```

```python
#create a new data frame with only unique rows of df with columns sd_id, sd_key, sd_name
new_df = pd.DataFrame(columns=["sd_id", "sd_key", "sd_name"])
new_df["sd_id"] = df["sd_id"].unique()
new_df["sd_key"] = df["sd_key"].unique()
new_df["sd_name"] = df["sd_name"].unique()
```

```python
#Add new column where for each sd_id, the corresponding value from d1 dictionary is added in the new column
#new_df["oid"] = new_df["sd_id"].apply(lambda x:{"oid" : d1[x]})
#new_df
```

```python
#Add new column where for each sd_id, the corresponding value from d1 dictionary is added in the new column
new_df["oid2"] = new_df["sd_id"].apply(lambda x: d1[x] if x in d1 else None)
new_df
```

```python
new_df.to_dict('records')
```

```python
dict_to_write= new_df.to_dict(orient ='records')
```

```python
#remove oid2 which has None from dict_to_write
for items in dict_to_write:
  if items["oid2"] is None:
    items.pop("oid2", None)
    print(items.keys())
```

```python
dict_to_write
```

```python
#write new_df to yaml file
import yaml
with open('my_yaml.yaml', 'w') as f:
    #yaml.dump(new_df.to_dict(orient ='records'), f, sort_keys=False)
    yaml.dump(dict_to_write, f, sort_keys=False)
```
